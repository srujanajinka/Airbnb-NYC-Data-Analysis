##***3.Data Wrangling***

##Data Wrangling Code

```python
# 1. Handle Missing Values
print("Handling missing values...\n")

# Check the number of missing values in each column
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Drop columns with more than 50% missing values
threshold = 0.5 * len(data)
data = data.dropna(thresh=threshold, axis=1)
print("\nColumns with more than 50% missing values removed.\n")

# Fill missing values with appropriate strategies
for column in data.columns:
    if data[column].dtype == 'object':  # For categorical columns
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:  # For numerical columns
        data[column].fillna(data[column].median(), inplace=True)

print("Remaining missing values:\n", data.isnull().sum())

# 2. Handle Duplicate Rows
print("\nHandling duplicate rows...\n")
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
if duplicates > 0:
    data = data.drop_duplicates()
    print("Duplicate rows removed.")

# 3. Convert Data Types
print("\nConverting data types...\n")
# Convert price to numeric if it's not already
if data['price'].dtype == 'object':
    data['price'] = pd.to_numeric(data['price'], errors='coerce')

# Convert date columns to datetime (if present)
date_columns = ['last_review']  # Replace with actual date columns in your dataset
for col in date_columns:
    if col in data.columns:
        data[col] = pd.to_datetime(data[col], errors='coerce')

print("Data type conversion completed.")

# 4. Handle Outliers
print("\nHandling outliers...\n")
# Example: Cap outliers in the 'price' column
q1 = data['price'].quantile(0.25)
q3 = data['price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data['price'] = data['price'].clip(lower=lower_bound, upper=upper_bound)

print("Outliers handled in the 'price' column.")

# 5. Standardize Column Names
print("\nStandardizing column names...\n")
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
print("Column names standardized:", data.columns)

# 6. Encode Categorical Variables
print("\nEncoding categorical variables...\n")
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    data[col] = data[col].astype('category').cat.codes

print("Categorical variables encoded.")

# 7. Final Check
print("\nFinal dataset shape:", data.shape)
print("Dataset preview:\n", data.head())
```
